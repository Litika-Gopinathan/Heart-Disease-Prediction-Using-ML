# Handling Duplicate data

```python
data_dup = data.duplicated().any()
data_dup
```

```python
data = data.drop_duplicates()
```

# Data processing

```python
cate_val = []
cont_val = []

for col in data.columns:
    if data[col].nunique() <= 10:
        cate_val.append(col)
    else:
        cont_val.append(col)
```

```python
cate_val
```

```python
cont_val
```

# Encoding Categorical data

```python
cate_val
```

```python
cate_val.remove('sex')
cate_val.remove('fbs')
cate_val.remove('restecg')
cate_val.remove('exang')
cate_val.remove('target')
```

```python
data = pd.get_dummies(data,columns=cate_val,drop_first=False)
```

```python
data.head()
```

```python
data.describe()
```

# Checking the distribution of target variable

```python
data['target'].value_counts()
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
data[cont_val] = st.fit_transform(data[cont_val])
```

```python
data.head()
```

# Split into training and testing dataset

```python
X = data.drop(columns='target', axis=1)
Y = data['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
```

```python
log=LogisticRegression()
log.fit(X_train,Y_train)
```

```python
ypred1=log.predict(X_test)
```

```python
unique_values, counts = np.unique(ypred1, return_counts=True)

# Print the results
for value, count in zip(unique_values, counts):
    print(f"{value} occurs {count} times")
```

```python
accuracy_score(Y_test,ypred1)
```

# SVC

```python
from sklearn import svm
```

```python
svm=svm.SVC()
```

```python
svm.fit(X_train,Y_train)
```

```python
ypred2=svm.predict(X_test)
```

```python
accuracy_score(Y_test,ypred2)
```
